```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
query = 'ChatGPT until:2023-01-10 since:2023-01-01'
limit = 100000
tweets_1 = []
for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    #print(vars(tweet))
    #break
    if len(tweets_1) == limit:
        break
    else:
        tweets_1.append([tweet.id,
                       tweet.user.username,
                       tweet.user.verified,
                       tweet.user.location,
                       tweet.user.followersCount,
                       tweet.rawContent,
                       tweet.date,
                       tweet.retweetedTweet,
                       tweet.lang, tweet.retweetCount])
df_1 = pd.DataFrame(tweets_1, columns = ['UserId',
                                     'UserName',
                                     'Verified',
                                     'Location',
                                     'Followers',
                                     'Tweet',
                                     'Date',
                                     'Retweeted',
                                     'Language', 'retweetCount'])
df_1.head()
df_1.to_csv('/content/drive/MyDrive/Sentiment_Analysis/extracted_files/ChatGPT_Sentiment/jan_2023.csv', index=False)
```

### Define a functin to loop through the folder and load the csv files containing certain defiend file name

```python
import os
import pandas as pd

def load_and_concat_data(folder_path, keyword, output_file):
    """
    Load and concatenate all datasets in a folder containing a specific keyword in the filename.

    Parameters:
    folder_path (str): Path to the folder containing the datasets.
    keyword (str): Keyword to search for in the filenames of the datasets to be loaded.
    output_file (str): Path to the output CSV file.

    Returns:
    None
    """
    # Find all files in the folder containing the keyword in the filename
    file_paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if keyword in f]

    # Load all files into a list of dataframes
    dfs = [pd.read_csv(file_path, low_memory=False) for file_path in file_paths]

    # Concatenate the dataframes into a single dataframe
    concatenated_df = pd.concat(dfs, ignore_index=True)

    # Save the concatenated dataframe to a CSV file
    concatenated_df.to_csv(output_file, index=False)

    print(f"Loaded and concatenated {len(dfs)} datasets. Saved to {output_file}")
```

```python
load_and_concat_data('/content/drive/MyDrive/Sentiment_Analysis/extracted_files/ChatGPT_Sentiment',
                     '2023',
                     '/content/drive/MyDrive/Sentiment_Analysis/extracted_files/ChatGPT_Sentiment/Chatgpt_sentiment.csv')
```
